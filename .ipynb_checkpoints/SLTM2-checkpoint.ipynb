```python
def get_trading_days(start_date, end_date):
    nyse = mcal.get_calendar('NYSE')
    schedule = nyse.schedule(start_date=start_date, end_date=end_date)
    trading_days = nyse.valid_days(start_date=schedule.index[0], end_date=schedule.index[-1])
    def makethisdate(date):
        new = date.date()
        new = str(new)
        new = pd.to_datetime(new)
        return new
    trading_days = list(map(makethisdate, trading_days))
    trading_days = pd.DataFrame(trading_days, columns=["dates"])
    trading_days['weekday'] = trading_days['dates'].dt.dayofweek
    trading_days["end_of_week"] = np.where(trading_days['dates'].dt.dayofweek == 4, True,False)
    trading_days["end_of_month"] = np.where(trading_days['dates'].dt.month != trading_days['dates'].shift(-1).dt.month, True,False)
    
    today = dt.date.today()
    csv_path = f"csv_files/tade_dates_{today}.csv"
    trading_days.to_csv(csv_path)
    
    return trading_days
```

```python
def get_trading_days(start_date, end_date):
    nyse = mcal.get_calendar('NYSE')
    schedule = nyse.schedule(start_date=start_date, end_date=end_date)
    trading_days = nyse.valid_days(start_date=schedule.index[0], end_date=schedule.index[-1])
    return trading_days

def generate_csv(start_date, end_date, output_file):
    trading_days = get_trading_days(start_date, end_date)
    
    with open(output_file, 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['Date', 'FirstDayOfWeek', 'FirstDayOfMonth'])
        
        prev_week = None
        prev_month = None
        
        for day in trading_days:
            date = day.strftime('%Y-%m-%d')
            
            week = day.weekday()
            first_day_of_week = prev_week != week
            prev_week = week
            
            month = day.month
            first_day_of_month = prev_month != month
            prev_month = month
            
            writer.writerow([date, first_day_of_week, first_day_of_month])


current_year = datetime.now().year
start_date = datetime(current_year, 1, 1)
```

```python
stocks = ['TSLA','XOM']
end1 = dt.date.today()
start1 = end1 - dt.timedelta(days=11*365)
start = start1.strftime('%Y-%m-%d')
end = end1.strftime('%Y-%m-%d')

stockdata = pdr.get_data_yahoo(stocks, start=start, end=end)
stockdata = stockdata['Adj Close']

returns = stockdata.pct_change().dropna(how='all').dropna(axis=1)
```

```python
# Import stock prices into temp table
temp_table = pd.DataFrame()
temp_table['price'] = stockdata[['TSLA']]

# Convert datetimes to dates
dates = temp_table['price'].index.tolist()
def makethisdate(date):
    new = date.date()
    new = str(new)
    new = pd.to_datetime(new)
    return new
dates = list(map(makethisdate, dates))
temp_table.index = dates

# Get list of months
month_df = pd.read_csv("days.csv")
month_df = pd.DataFrame(month_df)
month_df=month_df[month_df['FirstDayOfMonth']==True][['Date']]

# filter dates to only include end of month
temp_table = temp_table[temp_table.index.isin(month_df["Date"])]

display(temp_table.tail(10))
```

```python
# Packages related to machine learning
#for nueral networs
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dropout
import tensorflow.keras.backend as K

# fix random seed for same reproducibility as my results due to stochastic nature of start point
K.clear_session()
tf.keras.backend.clear_session()
np.random.seed(42)
tf.random.set_seed(42)
```

```python
sc = MinMaxScaler(feature_range=(0,1))
X = temp_table.drop(columns=['y'])
y = temp_table['y']

end1 = dt.date.today() - dt.timedelta(days=60)
start1 = end1 - dt.timedelta(days=5*365)
start = start1.strftime('%Y-%m-%d')
end = end1.strftime('%Y-%m-%d')

train = temp_table[start:end]
test = temp_table[end:]

scaler = sc.fit(train)
train_scaled = scaler.transform(train)
test_scaled = scaler.transform(test)

y_train_scaled = train_scaled[:,-0]
X_train_scaled = train_scaled[:,:9]
y_test_scaled = test_scaled[:,:-0]
X_test_scaled = test_scaled[:,:9]

y_train_scaled = y_train_scaled
X_train_scaled = X_train_scaled
y_test_scaled = y_test_scaled
X_test_scaled = X_test_scaled

X_train_scaled, X_test_scaled = np.array(X_train_scaled), np.array(X_test_scaled)
X_train_scaled = np.reshape(X_train_scaled, (X_train_scaled.shape[0], X_train_scaled.shape[1],1))
X_test_scaled = np.reshape(X_test_scaled, (X_test_scaled.shape[0], X_test_scaled.shape[1],1))

y_train_scaled, y_test_scaled = np.array(y_train_scaled), np.array(y_test_scaled)
```

```python
model = Sequential()
model.add(LSTM(units=50,return_sequences=True,input_shape=(X_train_scaled.shape[1], 1)))
model.add(Dropout(0.2))
model.add(LSTM(units=50,return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units=50,return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units=50))
model.add(Dropout(0.2))
model.add(Dense(units=1))
model.compile(optimizer='adam',loss='mean_squared_error')
model.fit(X_train_scaled,y_train_scaled,epochs=30,batch_size=68)
```

```python
predicted_stock_price = model.predict(X_test_scaled)
predicted_stock_price = sc.inverse_transform(predicted_stock_price)
```
